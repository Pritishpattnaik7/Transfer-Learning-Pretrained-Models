```python
from tensorflow import keras
```

```python
from keras.applications.vgg16 import VGG16  #importing vgg16
model = VGG16()   #loading the model
```

```python
model.summary()
```

```python
from keras.preprocessing.image import load_img
from keras.applications.imagenet_utils import preprocess_input
import numpy as np
import matplotlib.pyplot as plt

#load imagage from file
image = load_img('chulbuli.jpeg', target_size=(224, 224))

#convert the ima(ge into pixels to a numpy array
image = np.array(image)

#reshape the data for the model
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

#prepare the image for the vgg model
image = preprocess_input(image)

my_image = plt.imread('chulbuli.jpeg')
plt.imshow(my_image)
```

```python
#prediction
yhat = model.predict(image)  
```

```python
#convert the probabilites to class labels
from keras.applications.vgg16 import decode_predictions
label = decode_predictions(yhat)
label
```
